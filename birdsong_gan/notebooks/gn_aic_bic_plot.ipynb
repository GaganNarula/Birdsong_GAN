# what were the hidden states?

```python
hidden_state_sizes = [list(a.keys()) for a in criteria['AIC_full']]
```

```python
best_model_aic = criteria['best_model_aic']
best_model_bic = criteria['best_model_bic']
```

```python
best_order_aic = [int(hidden_state_sizes[d][k]) for d,k in enumerate(best_model_aic)]
best_order_bic = [int(hidden_state_sizes[d][k]) for d,k in enumerate(best_model_bic)]
```

```python
plt.figure(figsize=(12,6))
plt.plot(best_order_aic, '-or')
plt.plot(best_order_bic, '-ob')
plt.xlabel('days')
plt.ylabel('best model order')
plt.legend(['AIC','BIC'])
plt.title('bird r15y5 , nz = 8')
#plt.savefig('/home/songbird/Dropbox/Work/MDGAN_paper/Figures_for_paper/Final_figures/Figure3/r15y5_nz8_aic_bic.jpg', dpi=100,
#           format='jpg')
#plt.close()
```
